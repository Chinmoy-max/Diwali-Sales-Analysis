**Exploratory Data Analysis(EDA)**

**Gender**

```python
df.columns
```

```python
# seaborn countplot
x = sns.countplot(x ='Gender',data=df)
for bars in x.containers:
  x.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
#Total Amount spent by Genders
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x = 'Gender',y = 'Amount',data = sales_gen)
```

**Age**

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data = sales_age)
```

From the above graphs we can see that most of the most of the buyers are of age group between 26-35 yrs female

**State**


```python
df.columns
```

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x='State', y='Orders', data = sales_state)
```

```python
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x='State', y='Amount', data = sales_state)
```

**Marital Status**


```python
ax = sns.countplot(data = df, x = 'Marital_Status', hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#Marital status vs Amount
sales_marital = df.groupby(['Marital_Status','Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x='Marital_Status', y='Amount', data = sales_marital, hue='Gender')
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#Profession wise expend
sales_occu = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation', y='Amount', data = sales_occu)
```

**Product**

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = 'Product_Category', hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
sns.set(rc={'figure.figsize':(25,5)})
```

```python
sales_prod = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category', y='Amount', data = sales_prod)
```
